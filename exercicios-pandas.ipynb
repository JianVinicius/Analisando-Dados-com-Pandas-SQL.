### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
filtro = (df3["DATA FINAL"].dt.month == 8) & (df3["DATA FINAL"].dt.year == 2008)
df_gasolina_agosto_2008 = df3.loc[filtro]

preco_medio = df_gasolina_agosto_2008["PREÇO MÉDIO REVENDA"].mean()

print('O preço Médio da revenda da gasolina em agosto de 2008 é de',round(preco_medio, 2))
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
filtro = (df3["DATA FINAL"].dt.month == 5 ) & (df3["DATA FINAL"].dt.year == 2014) & (df3["ESTADO"] == 'SAO PAULO')
df_gasolina_sp = df3.loc[filtro]

preco_medio_sp = df_gasolina_sp["PREÇO MÉDIO REVENDA"].mean()

preco_medio_sp
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
preco_maximo_por_estado = df3.groupby("ESTADO").agg({"PREÇO MÉDIO REVENDA": "max"})
preco_maximo_por_estado


estados_acima_de_5_reais = preco_maximo_por_estado.loc[preco_maximo_por_estado["PREÇO MÉDIO REVENDA"] > 5.00]

if len(estados_acima_de_5_reais) == 0:
    print("Não foram encontrados estados com preço máximo acima de R$ 5,00.")
else:
    for estado in estados_acima_de_5_reais.index:
        filtro = (df_gasolina["ESTADO"] == estado) & (df_gasolina["PREÇO MÉDIO REVENDA"] > 5.00)
        datas = df_gasolina.loc[filtro, "DATA FINAL"]
        data_minima = datas.min()
        data_maxima = datas.max()
        print(f"No estado de {estado}, na data de {data_minima} atingiu valor minimo  e na data de  {data_maxima} atingiu valor maximo")
```

### 11. Qual a média de preço dos estados da região sul em 2012?

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
dfrj = df3[df3['ESTADO'] == 'RIO DE JANEIRO']

dfrj['ANO'] = dfrj['DATA FINAL'].dt.year

dfrj['VARIAÇÃO %'] = dfrj['PREÇO MÉDIO REVENDA'].pct_change()

tabela_variacao_rj = dfrj.groupby('ANO')['VARIAÇÃO %'].mean()

tabela_variacao_rj

```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
# Selecionando apenas a coluna PREÇO MÉDIO REVENDA e agrupando por estado

df_precos_estados = df3.groupby('ESTADO')['PREÇO MÉDIO REVENDA']

# Obtendo os preços máximo e mínimo registrados em cada estado

preco_max = df_precos_estados.max()
preco_min = df_precos_estados.min()

# Calculando a diferença absoluta e percentual entre os preços mais caros e mais baratos registrados em cada estado
dif_abs = preco_max - preco_min
dif_pct = (preco_max / preco_min - 1) * 100


# Obtendo os estados com os maiores e menores preços registrados
estado_maior_preco = dif_abs.idxmax()
estado_menor_preco = dif_abs.idxmin()


# Criando a tabela com a diferença absoluta e percentual entre os preços mais caros e mais baratos registrados em cada estado
tabela_precos = pd.concat([dif_abs, dif_pct], axis=1)
tabela_precos.columns = ['DIFERENÇA ABSOLUTA', 'DIFERENÇA PERCENTUAL']
tabela_precos['MAIOR PREÇO'] = preco_max
tabela_precos['MENOR PREÇO'] = preco_min
tabela_precos['ESTADO MAIOR PREÇO'] = estado_maior_preco
tabela_precos['ESTADO MENOR PREÇO'] = estado_menor_preco


print(tabela_precos)
```
